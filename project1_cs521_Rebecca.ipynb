KNN

```
def euclidean(u,v):
  if(len(u)!=len(v)):
    raise ValueError('vectors are not the same length')
  sum = 0
  for i in range(len(u)):
    sum += (u[i] - v[i])**2
  return np.sqrt(sum)

def manhattan(u,v):
  if(len(u)!=len(v)):
    raise ValueError('vectors are not the same length')
  sum = 0
  for i in range(len(u)):
    sum += abs(u[i] - v[i])
  return sum

def hamming(w,z):
  if(len(w)!=len(z)):
    raise ValueError('vectors are not the same length')
  sum = 0
  for i in range(len(w)):
    if not (w[i]==z[i]):
      sum+= abs(w[i]-z[i])    #not only +1 to take into account the weight put on the feature
  return sum
```

```
class KNN:

    def __init__(self, class_column, continuous_features, discrete_features=np.array([]), K=3, cont_dist_fn= euclidean):
        self.cont_dist_fn = cont_dist_fn
        self.K = K
        self.class_column = class_column
        self.continuous_features = continuous_features
        self.discrete_features = discrete_features
        return
    
    def fit(self, train_data, weights_features=np.array([1])):
        ''' Store the training data using this method as it is a lazy learner'''
        if np.all(weights_features == 1): #no weight
          weights_features = np.ones( np.shape(train_data)[1], dtype=int)
        self.train_data = train_data * weights_features
        self.weights_features = weights_features
        return self

    def predict(self, test_data):
      distances = np.zeros((len(test_data), len(self.train_data)), dtype=float)
      knns = np.zeros((len(test_data), self.K), dtype=int)
      predictions = np.zeros(len(test_data), dtype=int)
      #calculate distance between the training & test samples and returns an array of shape [num_test, num_train]
      for i in range(len(test_data)): 
        
        if not (self.continuous_features.size == 0):
          
          u = test_data[i, self.continuous_features]  #take only continuous features
          for j in range(len(self.train_data)):
            v = self.train_data[j, self.continuous_features] 
            distances[i][j] += self.cont_dist_fn(u,v)
        if not (self.discrete_features.size == 0):  #now take only discrete features
          w = test_data[i, self.discrete_features] 
          for j in range(len(self.train_data)):
            z = self.train_data[j, self.discrete_features] 
            distances[i][j] += hamming(w,z)

        knns[i,:] = np.argsort(distances[i])[:self.K]
        neighbours_class = np.zeros((self.K), dtype=int)
        for k in range(self.K):
          #store class of closest neighbours
          neighbours_class[k] = self.train_data[knns[i,k], self.class_column]
        predictions[i] = np.argmax(np.bincount(neighbours_class)) #take most frequent class
      return predictions
        


```

```
inds = np.random.permutation(Ncleaned)                                                     #generates an indices array from 0 to N-1 and permutes it 

#split the dataset into train and test 
train = cleaned_dataset1[inds[:478]] #70% data in train
test = cleaned_dataset1[inds[478:]]

accuracies = np.zeros(15, dtype=float) #test k=1 to 15
cont_features = np.arange(1,10)  #take out id and class columns

model = KNN(-1,cont_features,K=5)
model = model.fit(train)
predictions = model.predict(test * model.weights_features)
accuracy = np.sum(predictions == test[:,-1])/test.shape[0]
print(f'k = 5 \t accuracy for breast cancer is {accuracy*100:.1f}.')
print(f'total 2 : {np.sum(test[:,-1]==2)}')
print(f'total 4 : {np.sum(test[:,-1]==4)}')
tp=0
tn=0
fp=0
fn=0
for i in range(len(predictions)):
  if (predictions[i] == test[i,-1]):
    if(predictions[i]==2):
      tp+=1
    if(predictions[i]==4):
      tn+=1
  else:
    if(predictions[i]==2):
      fp+=1
    if(predictions[i]==4):
      fn+=1

confusion_matrix = [[tp, fp],[fn,tn]]
print(f'confusion matrix for k=5 \n {confusion_matrix[0]} \n {confusion_matrix[1]} ')

for k in range(15):
  model = KNN(-1,cont_features,K=k+1)
  model = model.fit(train)
  predictions = model.predict(test * model.weights_features)
  accuracy = np.sum(predictions == test[:,-1])/test.shape[0]
  accuracies[k] = accuracy

plt.plot(np.arange(1,16), accuracies)
plt.ylabel('accuracy')
plt.xlabel('K')
plt.show()

```
